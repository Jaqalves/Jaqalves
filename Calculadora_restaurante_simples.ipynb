```python

valor = float(input())
porc = float(input())
div_conta = int(input())

valor_gorjeta = (porc / 100) * valor
valor_total = valor + valor_gorjeta
valor_div = valor_total / div_conta

print("Bill Split Calculator")
print(f"Total (including tip): ${valor_total}")
print(f"Each person pays: ${valor_div}")
```
